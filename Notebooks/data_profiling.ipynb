```python
#Normalisation of product table
products_df = df[[
    'item_id',
    'category',
    'unit_price',
    'handling_cost_per_unit',
    'holding_cost_per_unit_day',
    'item_popularity_score'
]].copy()

products_df.head() 
```

```python
#Normalisation of Warehouse table
warehouses_df = df[[
    'storage_location_id',
    'zone',
    'layout_efficiency_score'
]].copy()

warehouses_df.head() 
```

```python
warehouses_df.shape
```

```python
warehouses_df.duplicated().sum()
```

```python
warehouses_df['storage_location_id'].nunique() 
```

```python
warehouses_clean = warehouses_df.drop_duplicates(subset=['storage_location_id']).copy()

warehouses_clean.shape 
```

```python
#create warehouseid 
warehouses_clean['warehouse_id'] = 'WH_' + warehouses_clean['storage_location_id'].astype(str)

warehouses_clean = warehouses_clean[['warehouse_id', 'storage_location_id', 'zone', 'layout_efficiency_score']]

warehouses_clean.head() 
```

```python
#Normalisation of Inventory table 
inventory_df = df[[
    'item_id',
    'storage_location_id',
    'stock_level',
    'reorder_point',
    'reorder_frequency_days',
    'lead_time_days',
    'last_restock_date',
    'stockout_count_last_month',
    'turnover_ratio'
]].copy()

inventory_df.head()
```

```python
#map storage location id with warehouse id
warehouse_mapping = dict(
    zip(warehouses_clean['storage_location_id'], warehouses_clean['warehouse_id'])
)

inventory_df['warehouse_id'] = inventory_df['storage_location_id'].map(warehouse_mapping)

inventory_df.head() 
```

```python
#check if mapping worked
inventory_df['warehouse_id'].isnull().sum()
```

```python
#Dropping the source column
inventory_df = inventory_df.drop(columns=['storage_location_id'])

inventory_df.head() 
```

```python
#Create a primary key
inventory_df = inventory_df.reset_index(drop=True)
inventory_df['inventory_id'] = inventory_df.index + 1 
```

```python
#Reorder columns 
inventory_df = inventory_df[[
    'inventory_id',
    'item_id',
    'warehouse_id',
    'stock_level',
    'reorder_point',
    'reorder_frequency_days',
    'lead_time_days',
    'last_restock_date',
    'stockout_count_last_month',
    'turnover_ratio'
]]

inventory_df.head()
```

```python
inventory_df['inventory_id'].nunique()
```

```python
inventory_df['last_restock_date'] = pd.to_datetime(inventory_df['last_restock_date'])
```

```python
#Normalise Operations metrics table
operations_df = df[[
    'item_id',
    'storage_location_id',
    'picking_time_seconds',
    'order_fulfillment_rate',
    'KPI_score'
]].copy()

operations_df.head() 
```

```python
#mapping the warehouse id again
operations_df['warehouse_id'] = operations_df['storage_location_id'].map(warehouse_mapping)

operations_df = operations_df.drop(columns=['storage_location_id'])

operations_df.head() 
```

```python
#Add primary key
operations_df = operations_df.reset_index(drop=True)
operations_df['operations_id'] = operations_df.index + 1 
```

```python
#Reorder columns
operations_df = operations_df[[
    'operations_id',
    'item_id',
    'warehouse_id',
    'picking_time_seconds',
    'order_fulfillment_rate',
    'KPI_score'
]]

operations_df.head() 
```
