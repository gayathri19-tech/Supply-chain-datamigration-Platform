```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

password = quote_plus("#####")

engine = create_engine(
    f"postgresql+psycopg2://postgres:{password}@127.0.0.1:5432/logimigrate_db"
)

with engine.connect() as conn:
    print("Connection successful!") 
```

```python
import pandas as pd

products_clean = pd.read_csv("../data/processed/products.csv")

products_clean.to_sql(
    "products",
    engine,
    if_exists="append",
    index=False
)

print("products data loaded successfully!") 
```

```python
#creating a one reusable function to load the remaining table
import pandas as pd

def load_table(file_path, table_name):
    df = pd.read_csv(file_path)
    
    df.to_sql(
        table_name,
        engine,
        if_exists="append",
        index=False
    )
    
    print(f"{table_name} loaded successfully! Rows: {len(df)}") 
```

```python
load_table("../data/processed/warehouses.csv", "warehouses")
load_table("../data/processed/inventory.csv", "inventory")
load_table("../data/processed/demand.csv", "demand")
load_table("../data/processed/operations.csv", "operations") 
```
